# Luhn Algorithm in Python – Card Number Validation

## Introduction

The Luhn Algorithm is a simple checksum formula used to validate identification numbers such as credit card numbers, debit card numbers, IMEI numbers, and more. It helps detect typing mistakes and invalid card numbers before processing payments.

This project implements the Luhn Algorithm using Python to verify whether a card number is valid or invalid.

---

## How the Algorithm Works

The algorithm follows these steps:

1. Reverse the card number.
2. Add all digits in odd positions directly.
3. Double each digit in even positions.
4. If a doubled value becomes greater than or equal to 10, split and add its digits.

   * Example:
     12 → 1 + 2 = 3
5. Add the results from odd and even positions.
6. If the final total is divisible by 10, the card number is valid.

Each odd-positioned digit is converted into an integer and added.

### 5. Process Even Positioned Digits

```python

```

This selects digits in even positions.

Each digit is doubled:

```python
number = int(digit) * 2
```

If the result is greater than or equal to 10:

```python
number = (number // 10) + (number % 10)
```

The digits are split and added.

Example:

```python
16 → 1 + 6 = 7
```

---

### 6. Calculate Final Total

```python
total = sum_of_odd_digits + sum_of_even_digits
```

---

### 7. Validate the Card Number

```python

```

If the total is divisible by 10, the card number is valid.

````python
def verify_card_number(card_number):
    sum_of_odd_digits = 0
    card_number_reversed = card_number[::-1]
    odd_digits = card_number_reversed[::2]

    for digit in odd_digits:
        sum_of_odd_digits += int(digit)

    sum_of_even_digits = 0
    even_digits = card_number_reversed[1::2]
    for digit in even_digits:
        number = int(digit) * 2
        if number >= 10:
            number = (number // 10) + (number % 10)
        sum_of_even_digits += number
    total = sum_of_odd_digits + sum_of_even_digits
    return total % 10 == 0

def main():
    card_number = '4111-1111-4555-1142'
    card_translation = str.maketrans({'-': '', ' ': ''})
    translated_card_number = card_number.translate(card_translation)

    if verify_card_number(translated_card_number):
        print('VALID!')
    else:
        print('INVALID!')

main()
````
