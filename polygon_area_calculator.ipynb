```python
class Rectangle:
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def set_width(self, width):
        self.width = width

    def set_height(self, height):
        self.height = height

    def get_area(self):
        return self.width * self.height

    def get_perimeter(self):
        return 2 * self.width + 2 * self.height

    def get_diagonal(self):
        return (self.width ** 2 + self.height ** 2) ** 0.5

    def get_picture(self):
        if self.width > 50 or self.height > 50:
            return "Too big for picture."

        # Creates a string of '*' for the width, adds a newline, and multiplies by the height
        return (("*" * self.width) + "\n") * self.height

    def get_amount_inside(self, shape):
        # Calculate how many times it fits horizontally and vertically (integer division)
        fit_width = self.width // shape.width
        fit_height = self.height // shape.height

        return fit_width * fit_height

    def __str__(self):
        return f"Rectangle(width={self.width}, height={self.height})"


class Square(Rectangle):
    def __init__(self, side):
        # Call the parent class (Rectangle) __init__ method
        super().__init__(side, side)

    def set_side(self, side):
        self.width = side
        self.height = side

    def set_width(self, width):
        # Overrides Rectangle's method to ensure it stays a square
        self.set_side(width)

    def set_height(self, height):
        # Overrides Rectangle's method to ensure it stays a square
        self.set_side(height)

    def __str__(self):
        return f"Square(side={self.width})"
```
