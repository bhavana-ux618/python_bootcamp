````python
##implementation of Dijkstra’s Shortest Path Algorithm using a weighted graph

---

# 1. The Graph Structure

```python
my_graph = {
    'A': [('B', 5), ('C', 3), ('E', 11)],
```

This is a **weighted graph** stored as a dictionary.

Each key is a node, and its value is a list of tuples:

```python
(destination, distance)
```

Example:

```python
('B', 5)
```

means:

* there is a connection from `A` to `B`
* the cost/distance is `5`

So:

```python
'A': [('B', 5), ('C', 3)]
```

means:

* A → B costs 5
* A → C costs 3

---

# 2. The Function

```python
def shortest_path(graph, start, target = ''):
```

Parameters:

* `graph` → the graph dictionary
* `start` → starting node
* `target` → optional destination node

If target is empty, it prints paths to ALL nodes.

---

# 3. Unvisited Nodes

```python
unvisited = list(graph)
```

Creates a list of all nodes that still need processing.

Initially:

```python
['A', 'B', 'C', 'D', 'E', 'F']
```

---

# 4. Distance Table

```python
distances = {
    node: 0 if node == start else float('inf')
    for node in graph
}
```

This creates shortest known distances.

If start is `'A'`:

```python
{
 'A': 0,
 'B': inf,
 'C': inf,
 'D': inf,
 'E': inf,
 'F': inf
}
```

Meaning:

* distance from A to A = 0
* everything else is initially unknown (`infinity`)

---

# 5. Path Storage

```python
paths = {node: [] for node in graph}
paths[start].append(start)
```

Stores the actual shortest route.

Initially:

```python
{
 'A': ['A'],
 'B': [],
 'C': [],
 ...
}
```

---

# 6. Main Algorithm Loop

```python
while unvisited:
```

Runs until every node is processed.

---

# 7. Choosing the Closest Node

```python
current = min(unvisited, key=distances.get)
```

This is the heart of Dijkstra’s algorithm.

It picks the unvisited node with the smallest known distance.

Example:

```python
current = 'A'
```

because A has distance 0.

---

# 8. Checking Neighbor Nodes

```python
for node, distance in graph[current]:
```

Looks at all neighbors of the current node.

For A:

```python
('B', 5)
('C', 3)
('E', 11)
```

---

# 9. Relaxation Step (IMPORTANT)

```python
if distance + distances[current] < distances[node]:
```

Checks:

> “Is this newly discovered path shorter?”

Example:

Current = A

Distance to A = 0

Edge A → C = 3

So:

```python
3 + 0 < infinity
```

TRUE ✅

Update shortest distance:

```python
distances['C'] = 3
```

This process is called **relaxation** in graph algorithms.

---

# 10. Updating Paths

```python
paths[node] = paths[current][:]
paths[node].append(node)
```

Copies the current path and adds the new node.

Example:

```python
['A'] + ['C']
```

becomes:

```python
['A', 'C']
```

---

# 11. Removing Processed Node

```python
unvisited.remove(current)
```

Once a node is finalized, remove it.

---

# 12. Printing Results

```python
print(f'\n{start}-{node} distance: {distances[node]}')
```

Shows:

* shortest distance
* shortest route

---

# Example Output

Starting from A:

```text
A-B distance: 4
Path: A -> C -> B
```

Why?

Instead of:

```text
A -> B = 5
```

the algorithm found a shorter route:

```text
A -> C -> B
3 + 1 = 4
```

---

# Final Shortest Paths from A

| Destination | Distance | Path          |
| ----------- | -------- | ------------- |
| B           | 4        | A → C → B     |
| C           | 3        | A → C         |
| D           | 4        | A → C → D     |
| E           | 8        | A → C → E     |
| F           | 6        | A → C → B → F |

---

# Time Complexity

Your implementation uses:

```python
min(unvisited, key=distances.get)
```

which scans all nodes each time.

Complexity:

```text
O(V²)
```

where:

* V = number of vertices

A faster version uses a **priority queue (heapq)**:

```text
O((V + E) log V)
```

---



✅ Uses real graph theory concepts
✅ Implements Dijkstra correctly
✅ Tracks both distances and paths
✅ Uses dictionaries + list comprehensions well
✅ Good understanding of weighted graphs

````

````python
my_graph = {
    'A': [('B', 5), ('C', 3), ('E', 11)],
    'B': [('A', 5), ('C', 1), ('F', 2)],
    'C': [('A', 3), ('B', 1), ('D', 1), ('E', 5)],
    'D': [('C',1 ), ('E', 9), ('F', 3)],
    'E': [('A', 11), ('C', 5), ('D', 9)],
    'F': [('B', 2), ('D', 3)]
}

def shortest_path(graph, start, target = ''):
    unvisited = list(graph)
    distances = {node: 0 if node == start else float('inf') for node in graph}
    paths = {node: [] for node in graph}
    paths[start].append(start)

    while unvisited:
        current = min(unvisited, key=distances.get)
        for node, distance in graph[current]:
            if distance + distances[current] < distances[node]:
                distances[node] = distance + distances[current]
                if paths[node] and paths[node][-1] == node:
                    paths[node] = paths[current][:]
                else:
                    paths[node].extend(paths[current])
                paths[node].append(node)
        unvisited.remove(current)

    targets_to_print = [target] if target else graph
    for node in targets_to_print:
        if node == start:
            continue
        print(f'\n{start}-{node} distance: {distances[node]}\nPath: {" -> ".join(paths[node])}')

    return distances, paths

shortest_path(my_graph, 'A')
````
