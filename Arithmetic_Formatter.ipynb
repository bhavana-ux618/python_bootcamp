````python
def arithmetic_arranger(problems, show_answers=False):
    # Rule: No more than 5 problems
    if len(problems) > 5:
        return 'Error: Too many problems.'

    top_row = []
    bottom_row = []
    dash_row = []
    answer_row = []

    for problem in problems:
        # Split the problem into its components
        parts = problem.split()
        operand1, operator, operand2 = parts[0], parts[1], parts[2]

        # Rule: Only + or - operators allowed
        if operator not in ('+', '-'):
            return "Error: Operator must be '+' or '-'."

        # Rule: Operands must contain only digits
        if not (operand1.isdigit() and operand2.isdigit()):
            return 'Error: Numbers must only contain digits.'

        # Rule: Operands cannot be more than 4 digits long
        if len(operand1) > 4 or len(operand2) > 4:
            return 'Error: Numbers cannot be more than four digits.'

        # Calculate width of the current problem
        width = max(len(operand1), len(operand2)) + 2

        # Format each line for the current problem
        top_row.append(operand1.rjust(width))
        bottom_row.append(operator + " " + operand2.rjust(width - 2))
        dash_row.append("-" * width)

        # Calculate the answer if show_answers is True
        if show_answers:
            if operator == '+':
                answer = str(int(operand1) + int(operand2))
            else:
                answer = str(int(operand1) - int(operand2))
            answer_row.append(answer.rjust(width))

    # Combine the formatted rows with 4 spaces between each problem
    arranged_problems = "    ".join(top_row) + "\n" + \
                        "    ".join(bottom_row) + "\n" + \
                        "    ".join(dash_row)

    # Append the answers row if requested
    if show_answers:
        arranged_problems += "\n" + "    ".join(answer_row)

    return arranged_problems
````
