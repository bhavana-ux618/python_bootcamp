````python
Tower of Hanoi project

---

# What  Project Does

This program solves the classic **Tower of Hanoi** puzzle using **recursion**.

We have 3 towers:

```python id="dxunko"
A = [5, 4, 3, 2, 1]
B = []
C = []
```

Goal:

```text id="2ms32m"
Move all disks from A → C
```

while following the rules:

1. Move only one disk at a time
2. Never place a larger disk on a smaller one
3. Use B as helper storage

---

# Initial Tower Setup

```python id="cfyv5p"
A = list(range(NUMBER_OF_DISKS, 0, -1))
```

Creates:

```python id="p6j7ua"
[5, 4, 3, 2, 1]
```

This means:

| Number | Disk Size |
| ------ | --------- |
| 5      | Largest   |
| 1      | Smallest  |

Python lists act like stacks here.

---

# Recursive Function

```python id="m8n0eq"
def move(n, source, auxiliary, target):
```

Parameters:

| Parameter   | Meaning           |
| ----------- | ----------------- |
| `n`         | number of disks   |
| `source`    | starting tower    |
| `auxiliary` | helper tower      |
| `target`    | destination tower |

---

# Base Case

```python id="u5cc34"
if n <= 0:
    return n
```

This stops recursion.

Without this, the function would call itself forever.

In recursion, this is called the **base case**.

---

# Recursive Step

This is the core logic:

```python id="vdstxj"
move(n - 1, source, target, auxiliary)
```

Meaning:

> Move smaller disks out of the way first.

---

# Moving the Largest Disk

```python id="t8u3x4"
target.append(source.pop())
```

This performs the actual move.

### `source.pop()`

Removes top disk from source tower.

### `target.append()`

Places disk on target tower.

This mimics real stack behavior.

---

# Showing Progress

```python id="o9s1t7"
print(A, B, C, '\n')
```

Displays tower states after every move.

Example:

```text id="jzjqja"
[5, 4, 3, 2] [] [1]
```

Meaning:

* disk 1 moved from A → C

---

# Final Recursive Step

```python id="jjh0lc"
move(n - 1, auxiliary, source, target)
```

Now move the smaller disks onto the large disk.

This completes the recursive pattern.

---

# The Recursive Strategy

To move `n` disks:

## Step 1

Move `n-1` disks away

```text id="6n24t9"
source → auxiliary
```

## Step 2

Move largest disk

```text id="iq20ww"
source → target
```

## Step 3

Move smaller disks back

```text id="uzd2x8"
auxiliary → target
```

This pattern repeats recursively.

---

# Mathematical Formula

Minimum moves required:

2^n - 1

For 5 disks:

```text id="8xgtam"
31 moves
```

---

# Time Complexity

Tower of Hanoi has exponential complexity:

O(2^n)

Meaning runtime doubles with every additional disk.

---

# One Small Improvement

You currently use:

```python id="2q9oq7"
return n
```

Usually in Tower of Hanoi we simply use:

```python id="1u1vrq"
return
```

because the returned value is never used.

So cleaner version:

```python id="c5w4w0"
if n <= 0:
    return
```

---

# What the Project Demonstrates

✅ Understanding recursion
✅ Stack operations (`append/pop`)
✅ Recursive decomposition
✅ Algorithm simulation
✅ Mathematical reasoning
✅ Debugging and indentation awareness
✅ Problem-solving skills

````
