```python
# Vector Mathematics System in Python

Developed a Python-based Vector Mathematics System using Object-Oriented Programming principles to perform advanced vector operations in 2D and 3D space. The project implements custom vector classes with support for mathematical operations through operator overloading, making vector calculations intuitive and efficient.

## Key Features

* Built reusable `R2Vector` and `R3Vector` classes for 2D and 3D vector representation.
* Implemented vector arithmetic operations including:

  * Vector addition
  * Vector subtraction
  * Scalar multiplication
  * Dot product
  * Cross product
* Added vector magnitude (norm) calculation using Euclidean distance formulas.
* Used operator overloading methods such as `__add__`, `__sub__`, `__mul__`, and comparison operators for clean mathematical syntax.
* Implemented inheritance where `R3Vector` extends `R2Vector` to promote code reusability and scalability.
* Added custom string representations with `__str__` and `__repr__` for readable outputs and debugging.
* Designed dynamic attribute handling using `vars()` and `getattr()` to make the implementation flexible and generic.

## Technologies Used

* Python
* Object-Oriented Programming (OOP)
* Operator Overloading
* Inheritance & Polymorphism
* Mathematical Computing

## Learning Outcomes

This project strengthened my understanding of:

* Python OOP concepts
* Class inheritance
* Special/magic methods
* Mathematical vector operations
* Writing scalable and reusable code
```

```python
class R2Vector:
    def __init__(self, *, x, y):
        self.x = x
        self.y = y

    def norm(self):
        return sum(val**2 for val in vars(self).values())**0.5

    def __str__(self):
        return str(tuple(getattr(self, i) for i in vars(self)))

    def __repr__(self):
        arg_list = [f'{key}={val}' for key, val in vars(self).items()]
        args = ', '.join(arg_list)
        return f'{self.__class__.__name__}({args})'

    def __add__(self, other):
        if type(self) != type(other):
            return NotImplemented
        kwargs = {i: getattr(self, i) + getattr(other, i) for i in vars(self)}
        return self.__class__(**kwargs)

    def __sub__(self, other):
        if type(self) != type(other):
            return NotImplemented
        kwargs = {i: getattr(self, i) - getattr(other, i) for i in vars(self)}
        return self.__class__(**kwargs)

    def __mul__(self, other):
        if type(other) in (int, float):
            kwargs = {i: getattr(self, i) * other for i in vars(self)}
            return self.__class__(**kwargs)
        elif type(self) == type(other):
            args = [getattr(self, i) * getattr(other, i) for i in vars(self)]
            return sum(args)
        return NotImplemented

    def __eq__(self, other):
        if type(self) != type(other):
            return NotImplemented
        return all(getattr(self, i) == getattr(other, i) for i in vars(self))

    def __ne__(self, other):
        return not self == other

    def __lt__(self, other):
        if type(self) != type(other):
            return NotImplemented
        return self.norm() < other.norm()

    def __gt__(self, other):
        if type(self) != type(other):
            return NotImplemented
        return self.norm() > other.norm()

    def __le__(self, other):
        return not self > other

    def __ge__(self, other):
        return not self < other

class R3Vector(R2Vector):
    def __init__(self, *, x, y, z):
        super().__init__(x=x, y=y)
        self.z = z

    def cross(self, other):
        if type(self) != type(other):
            return NotImplemented
        kwargs = {
            'x': self.y * other.z - self.z * other.y,
            'y': self.z * other.x - self.x * other.z,
            'z': self.x * other.y - self.y * other.x
        }

        return self.__class__(**kwargs)
v1 = R3Vector(x=2, y=3, z=1)
v2 = R3Vector(x=0.5, y=1.25, z=2)
print(f'v1 = {v1}')
print(f'v2 = {v2}')
v3 = v1 + v2
print(f'v1 + v2 = {v3}')
v4 = v1 - v2
print(f'v1 - v2 = {v4}')
v5 = v1 * v2
print(f'v1 * v2 = {v5}')
v6 = v1.cross(v2)
print(f'v1 x v2 = {v6}')
```
