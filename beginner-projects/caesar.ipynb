# Caesar Cipher Encryption Project Using Python

## Introduction

The Caesar Cipher is one of the oldest and simplest encryption techniques used in cryptography. It is a type of substitution cipher in which each letter in the plaintext is shifted by a fixed number of positions in the alphabet. This project demonstrates how basic encryption and decryption techniques can be implemented using Python programming.

The main objective of this project is to understand the fundamentals of cryptography, string manipulation, loops, and indexing concepts in Python. The program takes a user message as input and converts it into encrypted text using a specified shift value called an offset.

This project helps beginners understand how data security concepts work at a basic level and provides a foundation for learning advanced encryption algorithms used in cybersecurity and secure communications.

---

# Steps Followed in the Project

## Step 1: Define the Function

A function named `caesar()` is created with two parameters:

* `message` → the text entered by the user
* `offset` → the number of positions each letter should shift

## Step 2: Store the Alphabet

A string containing all lowercase alphabets from `a` to `z` is stored in a variable named `alphabet`.

## Step 3: Create an Empty String

An empty string called `encrypted_text` is initialized to store the encrypted output.

## Step 4: Traverse Each Character

A loop is used to go through every character in the message using:

```python
for char in message.lower():
```

## Step 5: Check for Spaces

If the character is a space, it is directly added to the encrypted text without modification.

## Step 6: Find Character Position

The program finds the index position of each character in the alphabet using:

```python
index = alphabet.find(char)
```

## Step 7: Shift the Character

The index is shifted using the offset value:

```python
new_index = (index + offset) % len(alphabet)
```

The modulo operator ensures the alphabet wraps around correctly.

## Step 8: Generate Encrypted Text

The shifted character is added to the encrypted text string.

## Step 9: Display the Output

Finally, the program prints:

* Original plaintext
* Encrypted ciphertext

````python
def caesar(message, offset):
    alphabet = 'abcdefghijklmnopqrstuvwxyz'
    encrypted_text = ''

    for char in message.lower():
        if char == ' ':
            encrypted_text += char
        else:
            index = alphabet.find(char)
            new_index = (index + offset) % len(alphabet)
            encrypted_text += alphabet[new_index]
    print('plain text:', message)
    print('encrypted text:', encrypted_text)
````
